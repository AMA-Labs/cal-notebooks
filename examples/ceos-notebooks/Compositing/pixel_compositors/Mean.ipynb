# Mean Compositor

#### Pixels

A pixel represents an observation at a specific time and location. A pixel is constituded of several measurements. 

For example, a landsat pixel is an observation of response intensities in in bands designated for `red`, `blue`, `green`, `near infrared (nir)` (among other designations along the em spectrum)  

#### Pixel Stack
A pixel stack is a series of observations over a range of time. 

#### Mean Composite

A `mean composite` takes all observations in a pixel stack and generates a mean value for all of it's measurements. 

For example. All values of the `red` measurement would be reduced to a single `mean_red` value.

A composite would be a new pixel composed of mean values for all measurement types.

For example:
> `[mean_red, mean_green, mean_blue, mean_nir]`

# Code

#### Imports

```python
import datacube
dc = datacube.Datacube(app = "Mean Pixel Compositor")

landsat_dataset = dc.load(latitude =(37.02, 37.12) ,
                          longitude =  (-76.55, -76.45) ,
                          time = ('2015-1-1', '2015-12-31'),
                          product = "landsat8_c2l2_sr",
                          measurements = ['red', 'green', 'blue', 'nir', 'pixel_qa'],
                          output_crs = 'EPSG:6933',
                          resolution = (-30,30),
                        dask_chunks = {'time':1}
                         ) 
```

#### Generate Cloud Mask

```python
from cloud_masking_utils import landsat_qa_clean_mask
cloud_mask = landsat_qa_clean_mask(landsat_dataset, platform="LANDSAT_8", 
                                   collection="c2", level="l2")
```

#### Cloud Composite

```python
def mean_composite(dataset, clean_mask):
    cleaned_dataset = dataset.where(clean_mask)
    composite = cleaned_dataset.mean('time', skipna=True)
    return composite
```

#### Execution

```python
composite = mean_composite(landsat_dataset,cloud_mask)
```

#### Plot

```python
import numpy as np
rgb_arr = np.dstack([composite.red.values,composite.green.values,composite.blue.values])
```
